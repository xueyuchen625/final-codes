```python
import problem_one as result_pro1
import matplotlib.pyplot as plt
```

```python
month_result = result_pro1.get_result()
```

```python
month, max_price, min_price, ave_price = result_pro1.get_x_and_y(month_result)
print(month)
```

```python
import numpy as np
total_width, n = 0.6, 3
width = total_width / n
x = np.arange(len(month))
x = x - (total_width - width)/2

plt.bar(x, max_price, width=width, label='', color='red')
plt.bar(x + width, min_price, width=width, label='', color='deepskyblue', tick_label=month)
plt.bar(x + width*2, ave_price, width=width, label='', color='green')

ave_show = []
for ele in ave_price:
    ave_show.append(int(ele))

for a, b in zip(x, max_price):
    plt.text(a, b+0.1, b, ha='center', va='bottom')
for a,b in zip(x, min_price):
    plt.text(a+width, b+0.1, b, ha='center', va='bottom')
for a,b in zip(x, ave_show):
    plt.text(a+2*width, b+0.1, b, ha='center', va='bottom')

plt.xticks()
plt.legend(loc="upper left")  # label
plt.rcParams['font.sans-serif'] = ['SimHei']  # 
plt.ylabel('')
plt.xlabel('')
plt.rcParams['savefig.dpi'] = 300  # 
plt.rcParams['figure.dpi'] = 300  # 
plt.rcParams['figure.figsize'] = (15.0, 8.0)  # 
plt.title("")
plt.savefig('F:\\result.png')

plt.show()
```
