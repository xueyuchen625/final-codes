```python
import problem_four as model
import matplotlib.pyplot as plt
```

```python
week_ave_dic = model.get_week_ave_price_dic()
```

```python
print(week_ave_dic)
```

```python
import numpy as np
week = []
ave_price = []
width = 0.2
for key in week_ave_dic:
    week.append(str(key)[2:])
    ave_price.append(round(week_ave_dic[key][0],1))

print(week)
print(ave_price)
x = np.arange(len(week))

plt.bar(x, ave_price, width = width, label="", color='deepskyblue', tick_label=week)
for a,b in zip(x, ave_price):
    plt.text(a, b+0.1, b, ha='center', va='bottom')

plt.legend(loc="upper left")
plt.rcParams['font.sans-serif'] = ['SimHei']  # 
plt.ylabel('')
plt.xlabel(' 171203 17123')
plt.rcParams['savefig.dpi'] = 300  # 
plt.rcParams['figure.dpi'] = 300  # 
plt.rcParams['figure.figsize'] = (30.0, 8.0)  # 
plt.title("")
plt.savefig('F:\\result_problem4.png')
plt.show()
```
